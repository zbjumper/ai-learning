列举常用数学函数图像

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, np.pi*2, 100)

sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x, sin_y, linestyle=':', label='sin')
plt.plot(x, cos_y, linestyle='--', label='cos')

plt.legend()

plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2, 1000)

y = x
y1 = x ** 2
y2 = np.sqrt(x)

plt.plot(x, y, label='y = x')
plt.plot(x, y1, label='y = x^2')
plt.plot(x, y2, label='y = sqrt(x)')

plt.legend()

plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
def sigmod(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 1000)
y = sigmod(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')

plt.grid()

plt.show()
```
